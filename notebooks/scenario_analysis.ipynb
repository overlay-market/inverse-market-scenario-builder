```python
import pandas as pd
import seaborn as sns
```

```python
path = '../csv/results.csv'
```

```python
df = pd.read_csv('../csv/results.csv').iloc[:, 1:]
mod_df = df.groupby(['ovl_amount', 'usdc_amount']).agg({'seconds': 'max'})
mod_df.reset_index(level='usdc_amount', inplace=True)
mod_df = mod_df.pivot(columns = 'usdc_amount')
mod_df.columns = mod_df.columns.droplevel()
mod_df.sort_index(ascending=False, inplace=True)
```

```python
mod_df
```

```python
sns.heatmap(mod_df, annot=True, fmt='.5g', cmap='viridis')
```
